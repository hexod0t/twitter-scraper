## Extract tweets from DataFrame

```python
texts = df['tweet']
texts
```

## Create a list with all the sentences

```python
all_sentences = []

for word in texts:
    all_sentences.append(word)
    
all_sentences
```

```python
lines = list()
for line in all_sentences:
    words = line.split()
    for w in words:
        lines.append(w)
        
print(lines)
```

## Preprocess Lines

```python
lines = [re.sub(r'[^A-Za-z0-9]+','',x) for x in lines]
filter_object = filter(lambda x: x != "", lines)
preprocessed_lines = list(filter_object)
```

```python
preprocessed_lines
```

```python
from nltk.stem.snowball import SnowballStemmer
s_stemmer = SnowballStemmer(language='english')

stem = []
for word in preprocessed_lines:
    stem.append(s_stemmer.stem(word))
    
stem
```

```python
stem2 = []
for word in stem:
    if word not in nlp.Defaults.stop_words:
        stem2.append(word)

stem2
```

```python
df = pd.DataFrame(stem2)
df = df[0].value_counts(0)
```

## Check Word Frequency

```python
from nltk.probability import FreqDist

freqdoctor = FreqDist()

for words in df:
    freqdoctor[words] += 1

freqdoctor
```
